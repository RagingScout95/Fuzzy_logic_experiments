```python
import numpy as np
import matplotlib.pyplot as plt
def trap(x,a,b,c,d,y):
    l = len(x)
    val=[]
    for i in range(l):
        if (x[i]<=a):
            val.append(0)
            
        elif (a<=x[i] and x[i]<=b):
            val.append(y)
            
        elif (b<=x[i] and x[i]<=c):
            val.append(y)
            
        elif (c<=x[i] and x[i]<=d):
            val.append((d-x[i])/(d-c))
            
        elif (x[i]>=d):
            val.append(0)
            
    return val
```

```python
temp = np.arange(0,10,1)
f1 = trap(temp,0, 1, 4, 5, 0.3)
f2 = trap(temp,3, 4, 6, 7, 0.5)
f3 = trap(temp,5, 6, 7, 8, 1)
fig,a=plt.subplots()
a.plot(temp, f1, 'r', temp, f2, 'g', temp, f3, 'k')
a.set_ylabel('uT ------>')
a.set_xlabel('T ------>')
a.set_ylim(0,1.5)

plt.text(2,0.35,'f1')
plt.text(4.5,0.55,'f2')
plt.text(6.2,1.05,'f3')
```

```python
t1 = np.fmax(f1,f2)
t2 = np.fmax(f2,f3)
t3 = np.fmax(t1,t2)
plt.plot(t3)
```

### Centroid Methods

```python
def cos(a,b,c,d,y):
    area = (((c-b)+(d-a))/2)*y
    cent = (a+d)/2
    l = [area,cent,y]
    return l
print("area centroid height")
A1 = cos(0, 1, 4, 5, 0.3)
A2 = cos(3, 4, 6, 7, 0.5)
A3 = cos(5, 6, 7, 8, 1)

print(A1)
print(A2)
print(A3)
a1 = A1[2]*A1[1] + A2[2]*A2[1] + A3[2]*A3[1]
b1 = A1[2] + A2[2]+ A3[2]
x1 = (A1[0]*A1[1] + A2[0]*A2[1] + A3[0]*A3[1])/(A1[0] + A2[0]+ A3[0])
x2 = (a1)/(b1)
print(a1)
print(b1)
print("Defuzzified output using COS :",format(x1))
print("Defuzzified output using weighted Average :",format(x2))
```

### Maxima Methods

```python
ymax = np.max(t3)

n = 0
P = []
for i in t3:
    if i == ymax:
        P.append(n)
    n=n+1

x = (P[0]+ P[-1])/2

plt.plot(t3)
ymax = np.max(t3)
plt.vlines(P, 0 , ymax,label = 'centroid', color='r')

print("The Defuzzified output using Mean of maxima is : {}".format(x))
```
