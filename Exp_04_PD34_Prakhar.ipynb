```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python
####========Complets Graphs========
g1 = nx.Graph()
n = input("Enter the value of n : ")
Node= []
for i in range(int(n)):
    Node.append(i)
g1.add_nodes_from(Node)
print("The number of nodes are : {}".format(g1.nodes))
#--------------------------
n = int(n)  # converting n into int 
for k in range(n):
    for m in range(n-1): # we want like n pair with 1,2,3... n-1
        if k==m: # we don't want (1,1), (2,2),.. like pairs and also it will not alter the graph but its not good practise
            continue
        else:
            ed = (k,m)
            g1.add_edge(*ed) #we use *ed because we r adding one edge at a time to add multibple edge we use add_edge_from(ed)
#--------------------------
print("The edges are : {}".format(g1.edges()))
A = nx.adjacency_matrix(g1)
print("The Adjacency matrix : ")
print(A.todense())
#--------------------------
nx.draw_networkx(g1)
plt.savefig("test.png")
plt.show()
```

```python
####========Cycle========
g2 = nx.Graph()
n = input("Enter the value of n : ")
Node= []
for i in range(int(n)):
    Node.append(i)
g2.add_nodes_from(Node)
print("The number of nodes are : {}".format(g2.nodes))
#--------------------------
l = 1 
n = int(n)
for k in range(n):
    if l==(n):
        l = 0
        ed = (k,l)
        g2.add_edge(*ed)
    else: # so its goes like k=0 pairs with l which is 1,2,3,..
        ed = (k,l) 
        l = l+1 # now here is the tricky part when k=n-2 then l will be n then in next loop we satisfied above condition 
        g2.add_edge(*ed)
#--------------------------
print("The edges are : {}".format(g2.edges()))
B = nx.adjacency_matrix(g2)
print("The Adjacency matrix : ")
print(B.todense())
#--------------------------
nx.draw_networkx(g2)
plt.savefig("test.png")
plt.show()
```

```python
####========Wheel Graphs========
g3 = nx.Graph()
n = input("Enter the value of n : ")
Node= []
for i in range(int(n)):
    Node.append(i)
g3.add_nodes_from(Node)
print("The number of nodes are : {}".format(g3.nodes))
#--------------------------
l = 1
N = int(n)
n = N+1 
for k in range(n): #everything is similar to wheel programe the only differece is we r making n as center so
    if l==(n):
        l = 0
        ed = (k-1,l)
#         print(ed)
        g3.add_edge(*ed)
        for m in range(n-1): # here we r pairing n with all other nodes
            ed = (k,m)
            g3.add_edge(*ed)
#             print(ed)
    else:
        ed = (k,l)
        l = l+1
        g3.add_edge(*ed)
#         print(ed)
#--------------------------
print("The edges are : {}".format(g3.edges()))
C = nx.adjacency_matrix(g3)
print("The Adjacency matrix : ")
print(C.todense())
#--------------------------
nx.draw_networkx(g3)
plt.savefig("test.png")
plt.show()
```

```python
####========Bipartite Graphs========
from networkx.algorithms import bipartite
g4 = nx.Graph() 
#---------------------------------------------------------------------------- 
a = input("Enter no. of nodes of set A : ") # taking input as n. of nodes in set A and asking user to specifei there names
N1= []
for i in range(int(a)):
    list1 = input("Enter Names for set A node {} : ".format(i))
    N1.append(list1)
g4.add_nodes_from(N1, bipartite=0)
print("---------------------------------------------")
b = input("Enter no. of nodes of set B : ")# taking input as n. of nodes in set B and asking user to specifei there names
N2= []
for j in range(int(b)):
    list2 = input("Enter Names for set B node {} : ".format(j))
    N2.append(list2)
g4.add_nodes_from(N2, bipartite=1)
#---------------------------------------------------------------------------- 
for k in range(int(a)): # so we r pairing it like complete graph as you can see its similar loop
    for m in range(int(b)):
        ed = (N1[k],N2[m])
        g4.add_edge(*ed)
#----------------------------------------------------------------------------           
B = nx.adjacency_matrix(g2)
print("The Adjacency matrix : ")
print(B.todense())
#---------------------------------------------------------------------------- 
nx.draw_networkx(g4, pos = nx.drawing.layout.bipartite_layout(g4, N1), width = 2)
```
