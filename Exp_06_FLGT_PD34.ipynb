```python
import networkx as nx
import numpy as np
G=nx.DiGraph()

G.add_node(1,pos=(0,0))
G.add_node(2,pos=(1,1))
G.add_node(3,pos=(1,-1))
G.add_node(4,pos=(2,0))
G.add_node(5,pos=(3,1))
G.add_node(6,pos=(3,-1))
G.add_node(7,pos=(4,0))

G.add_edge(1,2,weight=40)
G.add_edge(1,3,weight=40)
G.add_edge(2,4,weight=10)
G.add_edge(2,5,weight=10)
G.add_edge(3,4,weight=15)
G.add_edge(3,6,weight=20)
G.add_edge(4,5,weight=20)
G.add_edge(4,6,weight=15)
G.add_edge(4,7,weight=10)
G.add_edge(5,7,weight=30)
G.add_edge(6,7,weight=20)



pos=nx.get_node_attributes(G,'pos')
nx.draw_networkx(G,pos)
labels = nx.get_edge_attributes(G,'weight')
nx.draw_networkx_edge_labels(G,pos,edge_labels=labels)
```

```python
list1=[]
for path in nx.all_simple_paths(G, source=1, target=7):
    list1.append(path)
list1
```
