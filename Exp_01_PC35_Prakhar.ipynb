# -----User Defined Membership function----

## Triangular Membership function

```python
import numpy as np
import matplotlib.pyplot as plt

def tri(x,a,b,c):
    l = len(x)
    val=[]
    for i in range(l):
        if (x[i]<=a):
            val.append(0)
            
        elif (a<=x[i] and x[i]<=b):
            val.append((x[i]-a)/(b-a))
            
        elif (b<=x[i] and x[i]<=c):
            val.append((c-x[i])/(c-b))
            
        elif (x[i]>=c):
            val.append(0)
            
    return val
temp = np.arange(0,80,1)
hot = tri(temp,10, 20, 30)
moderate = tri(temp,30 , 40, 50)
cool = tri(temp,50, 60, 70)
fig,a=plt.subplots()
a.plot(temp, hot, 'r', temp, moderate, 'g', temp,cool, 'b')
a.set_ylabel('Fuzzy Membership')
a.set_xlabel('temp')
a.set_ylim(0,1.5)
```

## Trapezoidal Membership function

```python
import numpy as np
import matplotlib.pyplot as plt

def trap(x,a,b,c,d):
    l = len(x)
    val=[]
    for i in range(l):
        if (x[i]<=a):
            val.append(0)
            
        elif (a<=x[i] and x[i]<=b):
            val.append((x[i]-a)/(b-a))
            
        elif (b<=x[i] and x[i]<=c):
            val.append(1)
            
        elif (c<=x[i] and x[i]<=d):
            val.append((d-x[i])/(d-c))
            
        elif (x[i]>=d):
            val.append(0)
            
    return val
temp = np.arange(0,100,1)
hot = trap(temp,0, 10, 20, 30)
moderate = trap(temp,30 , 40, 50, 60)
cool = trap(temp,60, 70, 80, 90)
fig,a=plt.subplots()
a.plot(temp, hot, 'r', temp, moderate, 'g', temp,cool, 'b')
a.set_ylabel('Fuzzy Membership')
a.set_xlabel('temp')
a.set_ylim(0,1.5)
```

## Gaussian Membership function

```python
import numpy as np
import matplotlib.pyplot as plt
import math

def gauss(x,c,s):
    l = len(x)
    val=[]
    for i in range(l):
        val.append(math.exp((-1/2)*((x[i]-c)/s)**2))
    return val

temp = np.arange(0,22,1)
hot = gauss(temp,10,1)
moderate = gauss(temp,10,2)
cool = gauss(temp,10,3)
fig,a=plt.subplots()
a.plot(temp, hot, 'r', temp, moderate, 'g', temp,cool, 'b')
a.set_ylabel('Fuzzy Membership')
a.set_xlabel('temp')
a.set_ylim(0,1.5)
```

## Generalized bell Membership function

```python
import numpy as np
import matplotlib.pyplot as plt
import math

def gbell(x,a,b,c):
    l = len(x)
    val=[]
    for i in range(l):
        val.append(1/(1+((x[i]-c)/a)**(2*b)))
    return val

temp = np.arange(-10,10,1)
hot = gbell(temp,-3,3,0)
moderate = gbell(temp,-4,5,0)
cool = gbell(temp,-5,5,0)
fig,a=plt.subplots()
a.plot(temp, hot, 'r', temp, moderate, 'g', temp,cool, 'b')
a.set_ylabel('Fuzzy Membership')
a.set_xlabel('temp')
a.set_ylim(0,1.5)
```

## Sigmoid Membership function

```python
import numpy as np
import matplotlib.pyplot as plt
import math

def sig(x,a,c):
    l = len(x)
    val=[]
    for i in range(l):
        val.append(1/(1+math.exp(-a*(x[i]-c))))
    return val

temp = np.arange(0,10,1)
hot = sig(temp,1,5)
moderate = sig(temp,1,3)
cool = sig(temp,1,4)
fig,a=plt.subplots()
a.plot(temp, hot, 'r', temp, moderate, 'g', temp, cool, 'b')
a.set_ylabel('Fuzzy Membership')
a.set_xlabel('temp')
a.set_ylim(0,1.5)
```
